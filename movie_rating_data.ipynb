```python
from pandas import Series,DataFrame
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movie_data = pd.read_table('/home/bloombench/Documents/data-science/movielens/movies.dat',
               header=None,sep='::', names=['movie_id','name','genres'])
movie_data
```

```python
user_data = pd.read_table('/home/bloombench/Documents/data-science/movielens/users.dat',
               header=None,sep='::', names=['user_id','gender','age','occupation','zip_code'])
user_data
```

```python
rating_data = pd.read_table('/home/bloombench/Documents/data-science/movielens/ratings.dat',
               header=None,sep='::', names=['user_id','movie_id','rating','timestamp'])
rating_data
```

```python
data = pd.merge(pd.merge(rating_data,user_data),movie_data)
data
```

```python
mean_rating = data.pivot_table('rating',index='name',columns='gender', aggfunc='mean')
mean_rating
```

```python
#get top movies where rating >=250
grped_name = data.groupby('name').size()
top_movie_indexes = grped_name[grped_name>=250].index
top_movie_indexes
```

```python
# get mean ratings of only top movies
top_movie_mean_ratings = mean_rating.loc[top_movie_indexes]
top_movie_mean_ratings[:20].plot(kind='barh',stacked=True, figsize=(10,6))
plt.show()
```

```python
top_movie_mean_ratings['diff'] = top_movie_mean_ratings['F'] - top_movie_mean_ratings['M']
sorted_diff = top_movie_mean_ratings.sort_values(by='diff',ascending=False)
sorted_diff['diff'][:20].plot(figsize=(15,6))
plt.show()
```
