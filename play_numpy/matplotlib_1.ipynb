```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
marks = np.array([45,78,90,34,100,95])
plt.plot(marks) # by default x axis is array index
plt.show()
```

```python
plt.plot(marks,'om')
plt.show()
```

```python
physics = [94,96,89]
chemistry = [93,80,100]
names_pos = [0,1,2]
plt.xticks(names_pos,['abc','def','ghi'])
plt.plot(physics,'-r',label='Physics')
plt.plot(chemistry,'-b',label='Chemistry')
plt.legend(loc='upper left')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.show()
```

```python
p1 = plt.subplot(2,1,1)

physics = [94,96,89]
chemistry = [93,80,100]
p1.plot(physics,'-r',label='Physics')
p1.plot(chemistry,'-b',label='Chemistry')
plt.xlabel('Names')
plt.ylabel('Marks')
p1.legend(loc='best')
plt.xticks([0,1,2],['abc','def','ghi'])

p2 = plt.subplot(2,1,2)

quants = [45,87,34]
verbal = [97,50,80]
p2.plot(quants,'--r',label='Quants')
p2.plot(verbal,'--b',label='Verbal')
plt.xlabel('Names')
plt.ylabel('Marks')
p2.legend(loc='best')
plt.xticks([0,1,2],['abc','def','ghi'])



plt.show()
```
