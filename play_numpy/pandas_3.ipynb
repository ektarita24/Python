```python
data_list = []
data_1900 = pd.read_table('/home/bloombench/Documents/Ekta-Hadoop/hadoop-kit/data/weather_data_input/weather_data_set_1900',
               header=None,sep='|', names=['stationno','year','datemonth','hourminute','temperature'])
data_list.append(data_1900)
data_1900
```

```python
data_1901 = pd.read_table('/home/bloombench/Documents/Ekta-Hadoop/hadoop-kit/data/weather_data_input/weather_data_set_1901',
               header=None,sep='|', names=['stationno','year','datemonth','hourminute','temperature'])
data_list.append(data_1901)
data_1901
```

```python
#find the max temperature in year
weather_data = pd.concat(data_list,ignore_index=True)
cleaned_wdata = weather_data[weather_data['temperature'].notnull()]
projected_wdata = cleaned_wdata[['year','temperature']]
projected_wdata.groupby('year').max().plot(kind='bar',rot=0)
plt.show()
```
