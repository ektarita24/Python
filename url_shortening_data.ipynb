```python
path = '/home/bloombench/Documents/data-science/usagov_bitly_data2012-03-16-1331923249.txt'
open(path).readline()
```

```python
import json
lines = [json.loads(line) for line in open(path)]
lines
```

```python
from pandas import Series,DataFrame
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = DataFrame(lines)
data
```

```python
# get the frequency of url shortening taking place in a timezone
projected_data = data[['tz','u']]
cdata1 = projected_data['tz'].fillna('Unknown')
cdata1[cdata1 == ''] = 'Unknown'
tz_counts = cdata1.value_counts()
tz_counts[:10].plot(kind='barh')
plt.show()
```

```python
cdata2 = data[data['a'].notnull()]
cdata2['a']
```

```python
windows_or_non_windows = np.where(cdata2['a'].str.contains('Windows'),'Windows','Non Windows')
windows_or_non_windows
```

```python
grp1 = cdata2.groupby(['tz',windows_or_non_windows])
tz_oses = grp1.size().unstack().fillna(0)
tz_oses
```

```python
sort_indexes = tz_oses.sum(1).argsort()
result = tz_oses.take(sort_indexes)[-10:]
result.plot(kind = 'barh',stacked=True,figsize=(8,6))
plt.show()
```
