```python
from pandas import Series,DataFrame
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_table('/home/userapril17/interactions_of_type_category_output/part-r-00000', header=None,
                    sep = "\t",names = ['type','category','comments','likes','shares'])
data
```

```python
data['type_category'] = list(zip(data.type,data.category))
data
```

```python
projected_data = data[['type_category','comments','likes','shares']]
projected_data
```

```python
projected_data.plot(kind='barh', figsize=(15,10), stacked=True)
plt.yticks(np.arange(0,10),projected_data.type_category)
plt.show()
```
